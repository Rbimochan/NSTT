```python
import torch
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)
if DEVICE == "cuda":
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
CHECKPOINT_DIR = PROJECT_ROOT / "models" / "whisper-small-ne"
MAX_IN_DOMAIN = None   # None = full test manifest; set int for quick smoke
MAX_OOD = 50           # FLEURS samples (None = full test set)
INCLUDE_FLEURS = True

print("Checkpoint:", CHECKPOINT_DIR)
```

```python
from src.evaluation import run_evaluation

results = run_evaluation(
    PROJECT_ROOT,
    CHECKPOINT_DIR,
    device=DEVICE,
    max_in_domain=MAX_IN_DOMAIN,
    max_out_of_domain=MAX_OOD,
    include_fleurs=INCLUDE_FLEURS,
)

for r in results:
    print(f"{r.dataset}/{r.split}: WER={r.wer:.2%}, CER={r.cer:.2%} ({r.num_utterances} utts)")
```
